```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
#
x = np.linspace(-3,3,601)
y1 = np.zeros(len(x))
y2 = np.zeros(len(x))
y3 = np.zeros(len(x))
for i in range(len(x)):
    if x[i] <= 0.0:
        y1[i] = 1.0
    if x[i] <= 1.0:
        y2[i] = 1.0-x[i]
    y3[i] = math.log(1+math.exp(-x[i]))
```

```python
plt.plot(x,y1,'r-',lw=2)
plt.plot(x,y2,'b-',lw=2)
plt.plot(x,y3,'k-',lw=2)
plt.legend(['0-1', 'hinge', 'logistic'], fontsize=16)
plt.xlabel(r'$\mathrm{margin\ } y(w \cdot x)$', fontsize=18)
plt.ylabel(r'$\mathrm{loss}$', fontsize=18)
plt.xticks([-3,-2,-1,0,1,2,3],fontsize=16)
plt.yticks([0,1,2,3,4],fontsize=16)
plt.show()
```
